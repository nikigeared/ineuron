Create the below pattern using nested for loop in Python.

```python
for r in range(5):
    for c in range(r+1):
        print("*",end="")
    print()
for r in range(5,0,-1):
    for c in range(r-1):
        print("*",end="")
    print()
```

2. Write a Python program to reverse a word after accepting the input from the user.

```python
word=input("Enter a word:\n")
print(word[::-1])
```
